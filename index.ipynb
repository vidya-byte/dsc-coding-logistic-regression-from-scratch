## Gradient descent with the sigmoid function

Recall that gradient descent is a numerical method for finding a minimum to a cost function. In the case of logistic regression, you are looking to minimize the error between the model's predictions and the actual data labels. To do this, you first calculate an error vector based on the current model's feature weights. You then multiply the transpose of the training matrix itself by this error vector in order to obtain the gradient. Finally, you take the gradient, multiply it by the step size and add this to our current weight vector to update it. Below, write such a function. It will take 5 inputs:  

* `X`  
* `y`   
* `max_iterations`   
* `alpha` (the step size)   
* `initial_weights`   


By default, have your function set the `initial_weights` parameter to a vector where all feature weights are set to 1. 

```python
# Your code here
def grad_desc(X, y, max_iterations, alpha, initial_weights=None):
    if initial_weights is None:
        initial_weights = np.ones(X.shape[1])
    weights = initial_weights
    """Be sure to set default behavior for the initial_weights parameter."""
    # Create a for loop of iterations
    for iteration in range(max_iterations):
        predictions = sigmoid(np.dot(X, weights))
        error_vector = predictions - y
        # Generate predictions using the current feature weights
        # Calculate an error vector based on these initial predictions and the correct labels
        # Calculate the gradient
        gradient = np.dot(X.T, error_vector)
        # As we saw in the previous lab, calculating the gradient is often the most difficult task.
        # Here, your are provided with the closed form solution for the gradient of the log-loss function derived from MLE
        # For more details on the derivation, see the additional resources section below.
        gradient = np.dot(X.transpose(), error_vector) 
        # Update the weight vector take a step of alpha in direction of gradient
        weights -= alpha * gradient
    # Return finalized weights
    return(weights)
    
```

## Running your algorithm

Now that you've coded everything from the ground up, you can further investigate the convergence behavior of the gradient descent algorithm. Run your algorithm and plot the successive weights of the features through iterations. Below is a dataset, with `X` and `y` predefined for you. Use your logistic regression function to train a model. As the model trains, record the iteration cycle of the gradient descent algorithm and the weights of the various features. Then, plot this data on subplots for each of the individual features. Each graph should have the iteration number on the x-axis and the value of that feature weight for that iteration cycle on the y-axis. This will visually display how the algorithm is adjusting the weights over successive iterations, and hopefully show convergence to stable weights.

```python
# Your code here
def grad_desc(X, y, max_iterations, alpha, initial_weights=None):
    if initial_weights is None:
        initial_weights = np.ones(X.shape[1])  
    
    # Initialize weights
    weights = initial_weights
    
    # To store the weight updates at each iteration for plotting
    weights_history = []
    
    # Gradient Descent loop
    for iteration in range(max_iterations):
        predictions = sigmoid(np.dot(X, weights))
        
        # Calculate error vector (predictions - actual labels)
        error_vector = predictions - y
        
        # Calculate the gradient
        gradient = np.dot(X.T, error_vector)  
        
        # Update the weights by subtracting the gradient scaled by alpha
        weights -= alpha * gradient
        
        weights_history.append(weights)
    
    return weights, np.array(weights_history)
```

```python
# Create the predictor and target variables
y = df['target']
X = df.drop(columns=['target'], axis=1)

# Normalize X
X = (X - X.mean()) / X.std()

# Set hyperparameters
max_iterations = 1000
alpha = 0.01  # Learning rate

# Run gradient descent
final_weights, weights_history = grad_desc(X.values, y.values, max_iterations, alpha)

# Plot the weights over iterations for each feature
n_features = X.shape[1]
fig, axes = plt.subplots(n_features, 1, figsize=(10, 2 * n_features))

for i in range(n_features):
    axes[i].plot(range(max_iterations), weights_history[:, i])
    axes[i].set_title(f"Weight of Feature {X.columns[i]}")
    axes[i].set_xlabel("Iteration")
    axes[i].set_ylabel(f"Weight Value")
    axes[i].grid(True)

plt.tight_layout()
plt.show()
```

## Level up (Optional)

Update the gradient descent algorithm to also return the cost after each iteration. Then rerun the algorithm and create a graph displaying the cost versus the iteration number.

```python
# Your code here
def grad_desc(X, y, max_iterations, alpha, initial_weights=None):
    if initial_weights is None:
        initial_weights = np.ones(X.shape[1])  
    # Initialize weights
    weights = initial_weights
    
    # To store the weight updates at each iteration for plotting
    weights_history = []
    
    # To store the cost at each iteration for plotting
    cost_history = []
    
    # Gradient Descent loop
    for iteration in range(max_iterations):
        predictions = sigmoid(np.dot(X, weights))
        
        # Calculate error vector (predictions - actual labels)
        error_vector = predictions - y
        
        # Calculate the gradient
        gradient = np.dot(X.T, error_vector) 
        
        # Update the weights by subtracting the gradient scaled by alpha
        weights -= alpha * gradient
        
        # Store weights at this iteration for plotting
        weights_history.append(weights)
        
        # Calculate the cost (log-loss) and store it
        m = len(y)
        cost = - (1 / m) * np.sum(y * np.log(predictions) + (1 - y) * np.log(1 - predictions))
        cost_history.append(cost)
    
    return weights, np.array(weights_history), np.array(cost_history)
```

```python
# Run gradient descent with the updated function
final_weights, weights_history, cost_history = grad_desc(X.values, y.values, max_iterations, alpha)

# Plot the cost versus iteration number
plt.plot(range(max_iterations), cost_history)
plt.title('Cost vs Iteration')
plt.xlabel('Iteration')
plt.ylabel('Cost')
plt.grid(True)
plt.show()
```
